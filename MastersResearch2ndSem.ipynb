```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the main dataset
data_file_path = 'population-temp-gdp-4.csv'  # Update this path to your main CSV file
data = pd.read_csv(data_file_path)

# Handle missing values in the GINI Index using interpolation
data['GINI Coefficient'] = data['GINI Coefficient'].interpolate(method='linear', limit_direction='forward').fillna(method='bfill')

# Define features and target variable
features = data[['GDP (In USD)', 'Unemployment Rate (%)', 'Inflation Rate (%)', 'CO2 Emissions (In Tons)', 'Total Energy Consumption (KTOe)', 'Clean Water Access (%)', 'Population', 'GINI Coefficient', 'Human Development Index']]
target = data['Sustainable Development Goals Index']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Decision Tree Regressor
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
y_pred_dt = dt_regressor.predict(X_test)

# Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)

# Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_regressor.fit(X_train, y_train)
y_pred_gb = gb_regressor.predict(X_test)

# Evaluate models
dt_rmse = np.sqrt(mean_squared_error(y_test, y_pred_dt))
dt_r2 = r2_score(y_test, y_pred_dt)
rf_rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf))
rf_r2 = r2_score(y_test, y_pred_rf)
gb_rmse = np.sqrt(mean_squared_error(y_test, y_pred_gb))
gb_r2 = r2_score(y_test, y_pred_gb)

print("Decision Tree RMSE:", dt_rmse)
print("Decision Tree R2 Score:", dt_r2)
print("Random Forest RMSE:", rf_rmse)
print("Random Forest R2 Score:", rf_r2)
print("Gradient Boosting RMSE:", gb_rmse)
print("Gradient Boosting R2 Score:", gb_r2)

models = ['Decision Tree', 'Random Forest', 'Gradient Boosting']
rmse_scores = [dt_rmse, rf_rmse, gb_rmse]
r2_scores = [dt_r2, rf_r2, gb_r2]

# RMSE plot
plt.figure(figsize=(8, 6))
sns.barplot(x=models, y=rmse_scores)
plt.title('RMSE of Different Models')
plt.ylabel('RMSE')
plt.show()

# R2 Score plot
plt.figure(figsize=(8, 6))
sns.barplot(x=models, y=r2_scores)
plt.title('R2 Score of Different Models')
plt.ylabel('R2 Score')
plt.show()
```

```python
# -*- coding: utf-8 -*-
# Integrated: your original pipeline + time-aware cross-validation + matplotlib-only plots

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, TimeSeriesSplit, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score, make_scorer
import matplotlib.pyplot as plt

# -----------------------------
# 1) Load & prepare data
# -----------------------------
data_file_path = 'population-temp-gdp-4.csv'  # <-- update path if needed
data = pd.read_csv(data_file_path)

# Ensure chronological order if there's a Year column
# (Uncomment the next two lines if your CSV has a 'Year' column)
# if 'Year' in data.columns:
#     data = data.sort_values('Year').reset_index(drop=True)

# Handle missing values in GINI via interpolation + backfill (your original approach)
# data['GINI Coefficient'] = (
#     data['GINI Coefficient']
#     .interpolate(method='linear', limit_direction='forward')
#     .fillna(method='bfill')
# )

data = data.sort_values('Year').reset_index(drop=True)

data['GINI Coefficient'] = (
    data['GINI Coefficient']
    .interpolate(method='linear', limit_direction='forward')
    .fillna(method='bfill')
)
assert data['GINI Coefficient'].isna().sum() == 0


# Define features and target (your original set)
feature_cols = [
    'GDP (In USD)',
    'Unemployment Rate (%)',
    'Inflation Rate (%)',
    'CO2 Emissions (In Tons)',
    'Total Energy Consumption (KTOe)',
    'Clean Water Access (%)',
    'Population',
    'GINI Coefficient',
    'Human Development Index'
]
target_col = 'Sustainable Development Goals Index'

features = data[feature_cols]
target = data[target_col]

# -----------------------------
# 2) Hold-out split (your original approach)
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)

# -----------------------------
# 3) Fit models (your originals)
# -----------------------------
dt_regressor = DecisionTreeRegressor(random_state=42)
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)

dt_regressor.fit(X_train, y_train)
rf_regressor.fit(X_train, y_train)
gb_regressor.fit(X_train, y_train)

# Predictions (hold-out)
y_pred_dt = dt_regressor.predict(X_test)
y_pred_rf = rf_regressor.predict(X_test)
y_pred_gb = gb_regressor.predict(X_test)

# Evaluate (hold-out)
dt_rmse = np.sqrt(mean_squared_error(y_test, y_pred_dt))
dt_r2   = r2_score(y_test, y_pred_dt)

rf_rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf))
rf_r2   = r2_score(y_test, y_pred_rf)

gb_rmse = np.sqrt(mean_squared_error(y_test, y_pred_gb))
gb_r2   = r2_score(y_test, y_pred_gb)

print("=== HOLD-OUT RESULTS ===")
print(f"Decision Tree    -> RMSE: {dt_rmse:.6f} | R2: {dt_r2:.6f}")
print(f"Random Forest    -> RMSE: {rf_rmse:.6f} | R2: {rf_r2:.6f}")
print(f"Gradient Boosting-> RMSE: {gb_rmse:.6f} | R2: {gb_r2:.6f}")

# -----------------------------
# 4) Time-aware k-fold CV (k=5)
# -----------------------------
# Use chronological CV to avoid leakage for annual data
tscv = TimeSeriesSplit(n_splits=5)

def rmse_scorer_func(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
rmse_scorer = make_scorer(rmse_scorer_func, greater_is_better=False)

models = {
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=300, learning_rate=0.05, max_depth=3, random_state=42),
}

cv_results = {}
X = features.values
y = target.values

for name, model in models.items():
    r2_scores = cross_val_score(model, X, y, cv=tscv, scoring='r2')
    # For RMSE we invert the sign because make_scorer with greater_is_better=False returns negative values
    rmse_scores = -cross_val_score(model, X, y, cv=tscv, scoring=rmse_scorer)

    cv_results[name] = {
        "R2_mean":  float(np.mean(r2_scores)),
        "R2_sd":    float(np.std(r2_scores, ddof=1)),
        "RMSE_mean":float(np.mean(rmse_scores)),
        "RMSE_sd":  float(np.std(rmse_scores, ddof=1)),
        "R2_all":   r2_scores,
        "RMSE_all": rmse_scores,
    }

print("\n=== 5-FOLD TIME-SERIES CV RESULTS (mean  sd) ===")
for name, res in cv_results.items():
    print(f"{name:16s} R2: {res['R2_mean']:.3f}  {res['R2_sd']:.3f} | "
          f"RMSE: {res['RMSE_mean']:.3f}  {res['RMSE_sd']:.3f}")

# -----------------------------
# 5) Plots (matplotlib only)
# -----------------------------
model_names = ['Decision Tree', 'Random Forest', 'Gradient Boosting']
rmse_holdout = [dt_rmse, rf_rmse, gb_rmse]
r2_holdout   = [dt_r2,   rf_r2,   gb_r2]

# Hold-out RMSE
plt.figure(figsize=(8, 6))
plt.bar(model_names, rmse_holdout)
plt.title('Hold-out RMSE by Model')
plt.ylabel('RMSE')
plt.xlabel('Model')
plt.tight_layout()
plt.show()

# Hold-out R2
plt.figure(figsize=(8, 6))
plt.bar(model_names, r2_holdout)
plt.title('Hold-out R by Model')
plt.ylabel('R')
plt.xlabel('Model')
plt.tight_layout()
plt.show()

# CV RMSE (mean)
plt.figure(figsize=(8, 6))
plt.bar(model_names, [cv_results[m]['RMSE_mean'] for m in model_names])
plt.title('5-Fold Time-Series CV: RMSE (mean)')
plt.ylabel('RMSE')
plt.xlabel('Model')
plt.tight_layout()
plt.show()

# CV R2 (mean)
plt.figure(figsize=(8, 6))
plt.bar(model_names, [cv_results[m]['R2_mean'] for m in model_names])
plt.title('5-Fold Time-Series CV: R (mean)')
plt.ylabel('R')
plt.xlabel('Model')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
from tabulate import tabulate

# Feature Importance Analysis
def plot_feature_importance(model, feature_names):
    feature_importance = model.feature_importances_
    sorted_idx = np.argsort(feature_importance)
    plt.figure(figsize=(10, 8))
    plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
    plt.yticks(range(len(sorted_idx)), [feature_names[i] for i in sorted_idx])
    plt.xlabel('Feature Importance')
    plt.title('Feature Importance Analysis')
    plt.show()

    # Create a table of feature importances
    feature_importance_table = pd.DataFrame({
        'Feature': [feature_names[i] for i in sorted_idx],
        'Importance': feature_importance[sorted_idx]
    })

    # Print feature importance table
    print("\nFeature Importance Table:")
    print(tabulate(feature_importance_table, headers='keys', tablefmt='grid'))

# Plot feature importance for Decision Tree
print("\nDecision Tree Feature Importance:")
plot_feature_importance(dt_regressor, features.columns)

# Plot feature importance for Random Forest
print("\nRandom Forest Feature Importance:")
plot_feature_importance(rf_regressor, features.columns)

# Plot feature importance for Gradient Boosting
print("\nGradient Boosting Feature Importance:")
plot_feature_importance(gb_regressor, features.columns)
```

```python
# Support Vector Machine Regressor
svm_regressor = SVR()
svm_regressor.fit(X_train_scaled, y_train)
y_pred_svm = svm_regressor.predict(X_test_scaled)

# Evaluate the SVM model
mse_svm = mean_squared_error(y_test, y_pred_svm)
r2_svm = r2_score(y_test, y_pred_svm)
print(f"SVM - MSE: {mse_svm}, R2: {r2_svm}")
```

```python
# XGBoost Regressor
xgb_regressor = XGBRegressor(n_estimators=100, random_state=42)
xgb_regressor.fit(X_train, y_train)
y_pred_xgb = xgb_regressor.predict(X_test)

# Evaluate the XGBoost model
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)
print(f"XGBoost - MSE: {mse_xgb}, R2: {r2_xgb}")
```

```python
# Prepare data for LSTM
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# LSTM Model
lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
lstm_model.add(LSTM(units=50))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.fit(X_train_lstm, y_train, epochs=1000, batch_size=32, verbose=1)

# Predict and evaluate LSTM
y_pred_lstm = lstm_model.predict(X_test_lstm)
mse_lstm = mean_squared_error(y_test, y_pred_lstm)
r2_lstm = r2_score(y_test, y_pred_lstm)
print(f"LSTM - MSE: {mse_lstm}, R2: {r2_lstm}")
```

```python
# GRU Model
gru_model = Sequential()
gru_model.add(GRU(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
gru_model.add(GRU(units=50))
gru_model.add(Dense(1))
gru_model.compile(optimizer='adam', loss='mean_squared_error')
gru_model.fit(X_train_lstm, y_train, epochs=100, batch_size=32, verbose=1)

# Predict and evaluate GRU
y_pred_gru = gru_model.predict(X_test_lstm)
mse_gru = mean_squared_error(y_test, y_pred_gru)
r2_gru = r2_score(y_test, y_pred_gru)
print(f"GRU - MSE: {mse_gru}, R2: {r2_gru}")
```

```python
# Simple RNN Model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
rnn_model.add(SimpleRNN(units=50))
rnn_model.add(Dense(1))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
rnn_model.fit(X_train_lstm, y_train, epochs=100, batch_size=32, verbose=1)

# Predict and evaluate RNN
y_pred_rnn = rnn_model.predict(X_test_lstm)
mse_rnn = mean_squared_error(y_test, y_pred_rnn)
r2_rnn = r2_score(y_test, y_pred_rnn)
print(f"RNN - MSE: {mse_rnn}, R2: {r2_rnn}")
```

```python
# Extracting the years for the test set
test_years = data['Year'].iloc[-len(y_test):]

# Plotting actual vs predicted for the best model (assume Random Forest here)
plt.figure(figsize=(10, 6))
plt.plot(test_years, y_test, marker='o', linestyle='-', color='b', label='Actual Sustainability Index')
plt.plot(test_years, y_pred_rnn, marker='x', linestyle='--', color='r', label='Predicted Sustainability Index')
plt.xlabel('Year')
plt.ylabel('Sustainability Index')
plt.title('Actual vs Predicted Sustainability Index over Time (Random Forest)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, SimpleRNN
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.inspection import PartialDependenceDisplay
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

# Load the main dataset
data_file_path = 'population-temp-gdp-4.csv'
data = pd.read_csv(data_file_path)

# Handle missing values in the GINI Index using interpolation
data['GINI Coefficient'] = data['GINI Coefficient'].interpolate(method='linear', limit_direction='forward').fillna(method='bfill')

# Define features and target variable
features = data[['GDP (In USD)', 'Unemployment Rate (%)', 'Inflation Rate (%)', 'CO2 Emissions (In Tons)', 'Total Energy Consumption (KTOe)', 'Clean Water Access (%)', 'Population', 'GINI Coefficient', 'Human Development Index']]
target = data['Sustainable Development Goals Index']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42, shuffle=False)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Function to evaluate model
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{model_name} - MSE: {mse}, R2: {r2}")
    return y_pred, mse, r2

# Support Vector Machine Regressor with GridSearchCV
svm_regressor = SVR()
svm_param_grid = {'C': [0.1, 1, 10], 'gamma': [0.001, 0.01, 0.1], 'kernel': ['rbf']}
svm_grid = GridSearchCV(svm_regressor, svm_param_grid, cv=5)
svm_grid.fit(X_train_scaled, y_train)
y_pred_svm, mse_svm, r2_svm = evaluate_model(svm_grid.best_estimator_, X_test_scaled, y_test, "SVM")

# XGBoost Regressor with GridSearchCV
xgb_regressor = XGBRegressor(random_state=42)
xgb_param_grid = {'n_estimators': [50, 100, 150], 'learning_rate': [0.01, 0.1, 0.3], 'max_depth': [3, 5, 7]}
xgb_grid = GridSearchCV(xgb_regressor, xgb_param_grid, cv=5)
xgb_grid.fit(X_train, y_train)
y_pred_xgb, mse_xgb, r2_xgb = evaluate_model(xgb_grid.best_estimator_, X_test, y_test, "XGBoost")

# Prepare data for LSTM
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# LSTM Model
lstm_model = Sequential()
lstm_model.add(LSTM(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
lstm_model.add(LSTM(units=50))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.fit(X_train_lstm, y_train, epochs=100, batch_size=32, verbose=1)
y_pred_lstm, mse_lstm, r2_lstm = evaluate_model(lstm_model, X_test_lstm, y_test, "LSTM")

# GRU Model
gru_model = Sequential()
gru_model.add(GRU(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
gru_model.add(GRU(units=50))
gru_model.add(Dense(1))
gru_model.compile(optimizer='adam', loss='mean_squared_error')
gru_model.fit(X_train_lstm, y_train, epochs=100, batch_size=32, verbose=1)
y_pred_gru, mse_gru, r2_gru = evaluate_model(gru_model, X_test_lstm, y_test, "GRU")

# Simple RNN Model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(units=50, return_sequences=True, input_shape=(1, X_train_scaled.shape[1])))
rnn_model.add(SimpleRNN(units=50))
rnn_model.add(Dense(1))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')
rnn_model.fit(X_train_lstm, y_train, epochs=100, batch_size=32, verbose=1)
y_pred_rnn, mse_rnn, r2_rnn = evaluate_model(rnn_model, X_test_lstm, y_test, "RNN")

# Extracting the years for the test set
test_years = data['Year'].iloc[-len(y_test):]

# Plotting actual vs predicted for the best model (assume XGBoost here)
plt.figure(figsize=(10, 6))
plt.plot(test_years, y_test, marker='o', linestyle='-', color='b', label='Actual Sustainability Index')
plt.plot(test_years, y_pred_xgb, marker='x', linestyle='--', color='r', label='Predicted Sustainability Index')
plt.xlabel('Year')
plt.ylabel('Sustainability Index')
plt.title('Actual vs Predicted Sustainability Index over Time (XGBoost)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Partial Dependence Plots for XGBoost
features_to_plot = [0, 1, 2, 3]  # Indices of features to plot
fig, ax = plt.subplots(figsize=(12, 8))
display = PartialDependenceDisplay.from_estimator(xgb_grid.best_estimator_, X_train, features_to_plot, ax=ax)
plt.suptitle('Partial Dependence Plots for XGBoost Model')
plt.show()
```
